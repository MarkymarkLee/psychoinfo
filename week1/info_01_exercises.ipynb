## 1 Validating the findings about power poses (8 points)
Do <a href="https://www.youtube.com/watch?v=Ks-_Mh1QhMc">power poses</a> really work?
Let's analyze the data (keller2017.csv) from:

Keller, V. N., Johnson, D. J., & Harder, J. A. (2017). <a href="https://www.tandfonline.com/doi/abs/10.1080/23743603.2017.1341186">
Meeting your inner super (wo)man: are power poses effective when taught?</a> Comprehensive Results in Social Psychology, 2(1), 106-122.

where<br>

4 experimental conditions were AC (Aware-Constrictive), AE (Aware-Expansive), UC (Unaware-Constrictive), & UE (Unaware-Expansive)<br>
<br>
and<br>
<br>
3 dependent measures were body expansiveness, performance (in a job interview), & hireability scored by 4 independent raters.

Please construct 3 pandas dataframes df1, and df2, and df3 where <br><br>
<b>df1</b>, derived from df, has only 13 columns of interest (cond, expansiveness1-4, performance1-4, hire1-4) and no missing values (2 points) ;<br><br>
<b>df2</b>, derived from df1, has 4 columns that average expansiveness1-4 into "expansiveness", "performance1-4" into "performance", & hire1-4 into "hire" (df1 & df2 should have the same number of rows) (2 points) ; <br><br>
<b>df3</b>, derived from df2, calculates the means of each dependent variable in each condition (2 points) ;

Finally, run three two-way ANOVAs to see if there is any statistically significant result for each dependent measure (2 point). See <a href="https://www.statology.org/two-way-anova-python/">this page</a> for an example. 

```python
import pandas as pd
df=pd.read_csv("https://ceiba.ntu.edu.tw/course/4671ea/content/keller2017.csv")
df.head()
```

```python
df1 = df[["cond", "expansiveness_1", "expansiveness_2", "expansiveness_3", "expansiveness_4",'performance_1','performance_2','performance_3','performance_4','hire_1','hire_2','hire_3','hire_4']]
df1 = df1.dropna()
df1.head()
```

```python
expansiveness = df1[['expansiveness_1', 'expansiveness_2', 'expansiveness_3', 'expansiveness_4']].mean(axis=1)
performance = df1[['performance_1','performance_2','performance_3','performance_4']].mean(axis=1)
hire = df1[['hire_1','hire_2','hire_3','hire_4']].mean(axis=1)
df2 = pd.DataFrame({'cond':df1["cond"],'expansiveness':expansiveness, 'performance':performance, 'hire':hire})
df2.head()
```

```python
# AC, AE, UC, UE
ac_list = ["AC"] + list(df2[df2['cond'] == 'AC'][['expansiveness','performance','hire']].mean())
ae_list = ["AE"] + list(df2[df2['cond'] == 'AE'][['expansiveness','performance','hire']].mean())
uc_list = ["UC"] + list(df2[df2['cond'] == 'UC'][['expansiveness','performance','hire']].mean())
ue_list = ["UE"] + list(df2[df2['cond'] == 'UE'][['expansiveness','performance','hire']].mean())

df3 = pd.DataFrame([ac_list,ae_list,uc_list,ue_list], columns=['cond','expansiveness','performance','hire'])
df3
```

```python
print(df1.shape,df2.shape)
print(df3)
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

model = ols('expansiveness ~ performance * hire', data=df2).fit()
sm.stats.anova_lm(model, typ=2)
```

```python
model = ols('performance ~ hire * expansiveness', data=df2).fit()
sm.stats.anova_lm(model, typ=2)
```

```python
model = ols('hire ~ expansiveness * performance', data=df2).fit()
sm.stats.anova_lm(model, typ=2)
```

From the data above, we can observe that performance is the only factor that significantly affects hireability.
We can also observe that there is correlation between expansiveness and performance, but their relation is not as strong as performance and hireability.

However, there is quite likely that expansiveness does not affect hireability as the P-value is too large.
